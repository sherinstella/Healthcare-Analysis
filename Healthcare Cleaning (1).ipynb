```python

positive_data = cr1[cr1['Billing Amount'] > 0]['Billing Amount']

Q1 = positive_data.quantile(0.25)
Q3 = positive_data.quantile(0.75)
IQR = Q3 - Q1

valid_positive = positive_data[(positive_data >= Q1 - 1.5*IQR) & (positive_data <= Q3 + 1.5*IQR)]

median = valid_positive.median()   # or mean()



cr1["Billing Amount"]=cr1["Billing Amount"].fillna(median)

```
